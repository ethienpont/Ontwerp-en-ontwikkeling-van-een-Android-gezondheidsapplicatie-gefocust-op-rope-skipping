## Indelen data in frames


```python
import scipy.stats as stats
```

```python
Fs = 40 
frame_size = Fs*4 # 80
hop_size = Fs*4      #Fs*2 # 40
```

```python
def get_frames(df, frame_size, hop_size):

    N_FEATURES = 6

    frames = []
    labels = []
    for i in range(0, len(df) - frame_size, hop_size):
        x = df['acc_x'].values[i: i + frame_size]
        y = df['acc_y'].values[i: i + frame_size]
        z = df['acc_z'].values[i: i + frame_size]
        gx = df['gyro_x'].values[i: i + frame_size]
        gy = df['gyro_y'].values[i: i + frame_size]
        gz = df['gyro_z'].values[i: i + frame_size]
        # Retrieve the most often used label in this segment
        label = stats.mode(df['label'][i: i + frame_size])[0][0]
        frames.append([x, y, z,gx,gy,gz])
        labels.append(label)

    # Bring the segments into a better shape
    frames = np.asarray(frames).reshape(-1, frame_size, N_FEATURES)
    labels = np.asarray(labels)

    return frames, labels
```

```python
X, y = get_frames(Xnorm, frame_size, hop_size)
```

## CNN 2D model

```python
from keras.regularizers import l2
model = Sequential()
model.add(Conv2D(16, (3, 3), activation = 'relu', input_shape = X_train[0].shape))#3x1 - 32
model.add(BatchNormalization())
model.add(Dropout(0.1))
# extra conv
model.add(Conv2D(32, (3, 3), activation='relu'))# 16
model.add(Dropout(0.2))# na flatten

model.add(Flatten())

model.add(Dense(64, activation = 'relu', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))#bias eventueel weg
#model.add(Dropout(0.5))# eventueel weg

model.add(Dense(9, activation='softmax'))
```

```python
model.compile(optimizer=Adam(learning_rate = 0.00001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
#learning_rate verhogen
```

```python
model.summary()
```

```python
callback = []
checkpoint_path = "training_2/cp.ckpt"
callback.append(tf.keras.callbacks.ModelCheckpoint(
                    filepath=checkpoint_path,
                    monitor='val_loss',  # prediction
                    verbose=1,
                    save_best_only=True,
                    save_weights_only=True,  # TF 2.X
                    mode='min'            # prediction
                    # mode='max'              # classification
                    ))
```

```python
def plot_learningCurve(history, epochs):
  # Plot training & validation accuracy values
  epoch_range = range(1, epochs+1)
  plt.plot(epoch_range, history.history['accuracy'])
  plt.plot(epoch_range, history.history['val_accuracy'])
  plt.title('Model accuracy')
  plt.ylabel('Accuracy')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()

  # Plot training & validation loss values
  plt.plot(epoch_range, history.history['loss'])
  plt.plot(epoch_range, history.history['val_loss'])
  plt.title('Model loss')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend(['Train', 'Val'], loc='upper left')
  plt.show()
```
