```python
import pandas as pd
import numpy as np
```

```python
goal = 300
colnames = ["start", "end", "activity", "mets_score", "count"]
sessions = pd.read_csv(r"C:\Users\Elise\Documents\unif\master\semester2\masterproef\gitProject\thesis\data\recommendation_sample.csv", sep=';', names=colnames, header=None, skipinitialspace=True)
sessions
```

### gemiddeld aantal mets per int

```python
sessions['count'] = sessions.groupby('activity')['activity'].transform('count')
sessions
```

```python
sessions['duur'] = sessions['end'] - sessions['start']
sessions
```

```python
sessions['mets/int'] = sessions["mets_score"]/sessions["duur"]
sessions
```

```python
sessions["mean_mets/int"] = sessions.groupby('activity')['mets/int'].transform('mean')
sessions
```

```python
sessions["fractie"] = sessions["count"]/sessions.shape[0]
sessions
```

## sampling

```python
column_names = ["activity", "duur","mets"]
recommendations = pd.DataFrame(columns=column_names)
while (recommendations.sum()["mets"] < goal):
    act = np.random.choice(sessions["activity"])
    duration = np.random.uniform(low=0.2, high=3)
    mets = duration * sessions[sessions["activity"] == act]["mean_mets/int"].iloc[0]
    recommendations = recommendations.append(pd.Series(
        {
            "activity": act, "duur" : duration, "mets": mets
        }), ignore_index=True)
recommendations
```
